# use PyQt5 to create the graphic interface of "Who Wants to be a Programmer???"; a computer science trivia game

```python
import sys
from PyQt5.QtWidgets import QApplication, QLabel, QPushButton, QVBoxLayout, QWidget, QFileDialog, QGridLayout
from PyQt5.QtGui import QPixmap
from PyQt5 import QtGui, QtCore
from PyQt5.QtGui import QCursor
```

```python
#create global dictionary
widgets = { "logo": [],
            "button": [],
            "score": [],
            "question": [],
            "answer1": [],
            "answer2": [],
            "answer3": [],
            "answer4": []
         } 

app = QApplication(sys.argv) #initiallize GUI application
window = QWidget() #Create a window widget
window.setWindowTitle("Who wants to be a programmer???") #set window title
window.setFixedWidth(1000) #set window width
#window.move(2700, 200) #set window position along the monitor wit x & y cordinates
window.setStyleSheet("background: #161219;") #set background color using CSS code

grid = QGridLayout() #initialize grid layout

#hide all existing widgets and erase them from the global dictionary
def clear_widgets():
    for widget in widgets:
        if widgets[widget] != []:
            widgets[widget][-1].hide()
        for i in range(0, len(widgets[widget])):
            widgets[widget].pop()
    
def show_frame1():
    clear_widgets()
    frame1()
    
def start_game():
    #start the game, reset all widgets and parameters
    clear_widgets()
    #display the game frame
    frame2()
    
    
#create and design 4 buttons using function
def create_buttons(answer, l_margin, r_margin):
    button = QPushButton(answer)  
    button.setCursor(QCursor(QtCore.Qt.PointingHandCursor)) #set cursor to pointer
    button.setFixedWidth(485)
    button.setStyleSheet( 
        "*{border: 4px solid '#BC006C'; " +
        "margin-left:" + str(l_margin)+ "px;"+
        "margin-right:" + str(r_margin)+ "px;"+
        "border-radius: 25px; " +
        "font-family: Shanti; " +
        "font-size: 16px; "+
        "color: 'white';" +
        "padding: 15px 0;" +
        "margin-top: 20px;}" +
        "*:hover{background: '#BC006C';}"
    ) #styling include left and right margin parameters to button function
    button.clicked.connect(show_frame1)
    return button

#define frame functions as we will have many frames
def frame1():    
    #display logo
    image = QPixmap("logo.png") #load image file
    logo = QLabel() #create a label widget
    logo.setPixmap(image)  #place image inside label widget
    logo.setAlignment(QtCore.Qt.AlignCenter)  #align widget to center
    logo.setStyleSheet("margin-top: 100px;") #add mergin around the widget using CSS code
    widgets["logo"].append(logo)  #append local widgets logo into global dictionary

    #button widget
    button = QPushButton("PLAY") #create button widget
    button.setCursor(QCursor(QtCore.Qt.PointingHandCursor)) #set cursor to pointer
    button.setStyleSheet( 
        "*{border: 4px solid '#BC006C'; " +
        "border-radius: 45px; " +
        "font-size: 35px; "+
        "color: 'white';" +
        "padding: 25px 0;" +
        "margin: 100px 200px;}" +
        "*:hover{background: '#BC006C';}"
    )  #set border styling and styling for a mouse hover event #note: padding means space inside the element, margin means space outside the element in CSS
    
    button.clicked.connect(start_game) #button callback
    widgets["button"].append(button)  #append local widgets button into global dictionary   


    grid.addWidget(widgets["logo"][-1], 0, 0, 1, 2)  #place global widget inside a grid instructure 
    grid.addWidget(widgets["button"][-1], 1, 0, 1, 2)  #place global widget on the grid  
    

def frame2():
    score = QLabel("80")  #create a label widget and set the text value
    score.setAlignment(QtCore.Qt.AlignRight) #justify widget to the right side
    score.setStyleSheet( 
        "border: 1px solid '#64A314'; " +
        "border-radius: 35px; " +
        "font-size: 35px; "+
        "color: 'white';" +
        "padding: 15px 10px;" +
        "margin: 20px 200px;" +
        "background: '#64A314';" 
    )  #a quick styling
    
    widgets["score"].append(score)  #append local widgets score into global dictionary 
    
    question = QLabel("place holder text will go here blah blah!") #create a label widget
    question.setAlignment(QtCore.Qt.AlignCenter)  #align widget to center
    question.setWordWrap(True) #set the text wraping to True
    question.setStyleSheet( 
        "font-family: Shanti; " +
        "font-size: 35px; "+
        "color: 'white';" +
        "padding: 25px 15px;"
    )  #a quick styling
    
    widgets["question"].append(question)  #append local widgets question into global dictionary
    
    #create buttons inside frame2
    button1 = create_buttons("answer1", 85, 5) 
    button2 = create_buttons("answer2", 5, 85) 
    button3 = create_buttons("answer3", 85, 5) 
    button4 = create_buttons("answer4", 5, 85) 
    
    #append buttons to global dictionary
    widgets["answer1"].append(button1)
    widgets["answer2"].append(button2)
    widgets["answer3"].append(button3)
    widgets["answer4"].append(button4)
    
    #display logo_bottom
    image = QPixmap("logo_bottom.png") 
    logo = QLabel() 
    logo.setPixmap(image)
    logo.setAlignment(QtCore.Qt.AlignCenter) 
    logo.setStyleSheet("margin-top: 75px; margin-bottom: 30px;") 
    widgets["logo"].append(logo) 
    
    grid.addWidget(widgets["score"][-1], 0, 1)  #place global widget inside a grid instructure 
    grid.addWidget(widgets["question"][-1], 1, 0, 1, 2)  #place global widget on the grid, optional grid parameters to set the span:row, column
    grid.addWidget(widgets["answer1"][-1], 2, 0)
    grid.addWidget(widgets["answer2"][-1], 2, 1)
    grid.addWidget(widgets["answer3"][-1], 3, 0)
    grid.addWidget(widgets["answer4"][-1], 3, 1)
    grid.addWidget(widgets["logo"][-1], 4, 0, 1, 2)
    

frame2()

    
window.setLayout(grid) #aply grid on window object

window.show() #display window
sys.exit(app.exec()) #terminate the app
```
